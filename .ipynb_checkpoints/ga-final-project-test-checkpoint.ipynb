```python
import numpy as np

temp_max = np.array([17,  19,  21,  28,  33,  38, 37,  37,  31,  23,  19,  18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
```

```python
import matplotlib.pyplot as plt
months = np.arange(12)
plt.plot(months, temp_max, 'ro')
plt.plot(months, temp_min, 'bo')
plt.xlabel('Month')
plt.ylabel('Min and Max Temps')
```

```python
from scipy import optimize
# Defining the function to fit
# The first argument must be the independent variable
# The remaining arguments are parameters to fit
def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months, temp_max, [20,10,0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months, temp_min, [-40,20,0])
```

```python
days = np.linspace(0,12,num=365)

plt.figure()
plt.plot(months,temp_max,'ro')
plt.plot(days, yearly_temps(days, *res_max),'r-')
plt.plot(months,temp_min,'bo')
plt.plot(days, yearly_temps(days, *res_min),'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

```python
# Create periodic function to fit concentration curve
# a = amplitude
# s = horizontal shift
# p = period
# t = vertical shift

#def conc_curve(cum_vol, a1, s1, p1, a2, s2, p2, t):
#    return a1*np.sin(p1*(cum_vol-s1))+a2*np.cos(p2*(cum_vol-s2))+t

#def conc_curve(cum_vol, amp1, offset1, peaks, avg):
#    return avg + amp1*np.sin(2*np.pi/(cum_vol.max()/peaks)*(cum_vol+offset1))

def conc_curve(x,a1,a2,a3,a4):
    return (a1*x**3+a2*x**2+a3*x+a4)

x = [15.8,857.2,1422.3,1852.7,2383.1,3108.1,3667.9,4248.6,5585.9,6592.1,7600.7,8093.4,8486.7,8647.1,9009.1,9337.1,9670.1,10084.1,10215.2,11104.2,11922.2,12377]
y = [0,0,0.5,0.75,1,1.5,1.75,0,0.5,1,1.5,2,0,0.25,0,0.5,1,1.5,2,1.5,1.75,0]

#a = max(y)-min(y)
#t = 0

res_conc, cov_conc = optimize.curve_fit(conc_curve, x, y, [2,2,2,0])
```
